```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
#defining a dataframe of all the months 
columns_name=['MONTH','TOTAL_TRAFFIC']
dfresult=pd.DataFrame(columns=columns_name)
###############################
#defining a dataframe of each month 
column_name=['DATE','TOTAL_TRAFFIC']
dfoctresult=pd.DataFrame(columns=column_name)
dfsepresult=pd.DataFrame(columns=column_name)
dfaugresult=pd.DataFrame(columns=column_name)
dfjulresult=pd.DataFrame(columns=column_name)
dfjunresult=pd.DataFrame(columns=column_name)
dfmayresult=pd.DataFrame(columns=column_name)
dfaprresult=pd.DataFrame(columns=column_name)
dfmarresult=pd.DataFrame(columns=column_name)
dffebresult=pd.DataFrame(columns=column_name)
dfjanresult=pd.DataFrame(columns=column_name)
dfdecresult=pd.DataFrame(columns=column_name)
dfnovresult=pd.DataFrame(columns=column_name)
dfoct19result=pd.DataFrame(columns=column_name)
dfsep19result=pd.DataFrame(columns=column_name)
###############################
#reading the csv files and store them in dataframes
#then merging them into one dataframe
df1 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_201031.txt')
df2 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_201024.txt')
df3 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_201017.txt')
df4 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_201010.txt')
df5 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_201003.txt')
dfoct=[df1,df2,df3,df4,df5]
dfoct2020=pd.concat(dfoct,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df6 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200926.txt')
df7 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200919.txt')
df8 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200912.txt')
df9 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200905.txt')
dfsep=[df6,df7,df8,df9]
dfsep2020=pd.concat(dfsep,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df10 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200829.txt')
df11 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200822.txt')
df12 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200815.txt')
df13 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200808.txt')
df14 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200801.txt')
dfaug=[df10,df11,df12,df13,df14]
dfaug2020=pd.concat(dfaug,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df15 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200725.txt')
df16 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200718.txt')
df17 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200711.txt')
df18 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200704.txt')
dfjul=[df15,df16,df17,df18]
dfjul2020=pd.concat(dfjul,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df19 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200627.txt')
df20 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200620.txt')
df21 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200613.txt')
df22 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200606.txt')
dfjun=[df19,df20,df21,df22]
dfjun2020=pd.concat(dfjun,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df23 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200530.txt')
df24 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200523.txt')
df25 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200516.txt')
df26 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200509.txt')
df27 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200502.txt')
dfmay=[df23,df24,df25,df26,df27]
dfmay2020=pd.concat(dfmay,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
################################################################################################
dfapr1 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200425.txt')
dfapr2 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200418.txt')
dfapr3 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200411.txt')
dfapr4 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200404.txt')
dfapr=[dfapr1,dfapr2,dfapr3,dfapr4]
dfapr2020=pd.concat(dfapr,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
########################################lockdown################################################
dflock1 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200328.txt')
dflock2 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200321.txt')
dflock3 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200314.txt')
dflock4 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200307.txt')
dfmar=[dflock1,dflock2,dflock3,dflock4]
dfmar2020=pd.concat(dfmar,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
################################################################################################
df28 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200229.txt')
df29 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200222.txt')
df30 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200215.txt')
df31 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200208.txt')
df32 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200201.txt')
dffeb=[df28,df29,df30,df31,df32]
dffeb2020=pd.concat(dffeb,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df33 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200125.txt')
df34 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200118.txt')
df35 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200111.txt')
df36 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_200104.txt')
dfjan=[df33,df34,df35,df36]
dfjan2020=pd.concat(dfjan,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df37 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191228.txt')
df38 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191221.txt')
df39 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191214.txt')
df40 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191207.txt')
dfdec=[df37,df38,df39,df40]
dfdec2019=pd.concat(dfdec,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df41 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191130.txt')
df42 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191123.txt')
df43 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191116.txt')
df44 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191109.txt')
df45 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191102.txt')
dfnov=[df41,df42,df43,df44,df45]
dfnov2019=pd.concat(dfnov,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df46 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191026.txt')
df47 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191019.txt')
df48 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191012.txt')
df49 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_191005.txt')
dfoct19=[df46,df47,df48,df49]
dfoct2019=pd.concat(dfoct19,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
#*******************************************************************************************
df50 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_190928.txt')
df51 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_190921.txt')
df52 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_190914.txt')
df53 = pd.read_csv('http://web.mta.info/developers/data/nyct/turnstile/turnstile_190907.txt')
dfsep19=[df50,df51,df52,df53]
dfsep2019=pd.concat(dfsep19,keys=['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'])
# ########################################################
#this function to clean the dataframe, add new columns 
def cleanning(df):
    #to remove the space after the exit column
    df.columns=df.columns.str.strip()
    #drop unnecessary columns 
    df.drop(['C/A','UNIT','SCP','LINENAME','DIVISION','DESC'],axis=1,inplace= True)
    #converting to the standard date format 
    df["DATE"] = pd.to_datetime(df["DATE"])
    #calculating the actual entries not the cumulative 
    df['TOTAL_ENTRIES']= df['ENTRIES'].diff().tolist();
    #calculating the actual exits not the cumulative 
    df['TOTAL_EXITS']= df['EXITS'].diff().tolist();
    # drop the negative values
    df.drop(df.index[df['TOTAL_ENTRIES'] < 0.0 ], inplace = True)
    df.drop(df.index[df['TOTAL_EXITS'] < 0.0 ], inplace = True)
    #drop the outliers 60*60*24
    threshold=86400
    df.drop(df.index[df['TOTAL_ENTRIES'] > threshold],inplace=True )
    df.drop(df.index[df['TOTAL_EXITS'] > threshold],inplace=True )
    threshold = 100000000
    #calculating the total traffic
    df['TOTAL_TRAFFIC']=df['TOTAL_ENTRIES']+df['TOTAL_EXITS']
    #drop the outliers
    df.drop(df.index[df['TOTAL_TRAFFIC'] > threshold],inplace=True )
    #drop the nan values
    df.dropna(inplace= True)
    return df
#this function calculates the sum of the traffic in all the days in one month
def print_result(df):
    result=df['TOTAL_TRAFFIC'].sum().item()
    return result
#this function calculates the sum of the traffic in each day
def print_result_of_month(df):
    result=df.groupby(['DATE'])['TOTAL_TRAFFIC'].sum().astype(int).reset_index()
    return result
#1- cleanning
#2- get the values of days in one month
#3- write the values in a csv file
#4- get the total value in one month
#5- save the value in dictionary
#6- insert the dictionary into the result dataframe, which includes all the months
#7- create a csv file with all the months 
cleanning(dfoct2020)
print(dfoct2020)
dfoctresult=print_result_of_month(dfoct2020)
dfoctresult.to_csv("cleaned_data_oct.csv")
x={'MONTH':'OCTOBER 2020','TOTAL_TRAFFIC':print_result(dfoctresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfsep2020=cleanning(dfsep2020)
print(dfsep2020)
dfsepresult=print_result_of_month(dfsep2020)
dfsepresult.to_csv("cleaned_data_sep.csv")
x={'MONTH':'SEPTEMBER 2020','TOTAL_TRAFFIC':print_result(dfsepresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfaug2020=cleanning(dfaug2020)
print(dfaug2020)
dfaugresult=print_result_of_month(dfaug2020)
dfaugresult.to_csv("cleaned_data_aug.csv")
x={'MONTH':'AUGUST 2020','TOTAL_TRAFFIC':print_result(dfaugresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfjul2020=cleanning(dfjul2020)
print(dfjul2020)
dfjulresult=print_result_of_month(dfjul2020)
dfjulresult.to_csv("cleaned_data_jul.csv")
x={'MONTH':'JULY 2020','TOTAL_TRAFFIC':print_result(dfjulresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfjun2020=cleanning(dfjun2020)
print(dfjun2020)
dfjunresult=print_result_of_month(dfjun2020)
dfjunresult.to_csv("cleaned_data_jun.csv")
x={'MONTH':'JUNE 2020','TOTAL_TRAFFIC':print_result(dfjunresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfmay2020=cleanning(dfmay2020)
print(dfmay2020)
dfmayresult=print_result_of_month(dfmay2020)
dfmayresult.to_csv("cleaned_data_may.csv")
x={'MONTH':'MAY 2020','TOTAL_TRAFFIC':print_result(dfmayresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfapr2020=cleanning(dfapr2020)
print(dfapr2020)
dfaprresult=print_result_of_month(dfapr2020)
dfaprresult.to_csv("cleaned_data_apr.csv")
x={'MONTH':'APRIL 2020','TOTAL_TRAFFIC':print_result(dfaprresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
########################################################################
dfmar2020=cleanning(dfmar2020)
print(dfmar2020)
dfmarresult=print_result_of_month(dfmar2020)
dfmarresult.to_csv("cleaned_data_mar.csv")
x={'MONTH':'MARCH 2020','TOTAL_TRAFFIC':print_result(dfmarresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# ########################################################################
dffeb2020=cleanning(dffeb2020)
print(dffeb2020)
dffebresult=print_result_of_month(dffeb2020)
dffebresult.to_csv("cleaned_data_feb.csv")
x={'MONTH':'FEBRUARY 2020','TOTAL_TRAFFIC':print_result(dffebresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
#-------------------------------------------------------------------
dfjan2020=cleanning(dfjan2020)
print(dfjan2020)
dfjanresult=print_result_of_month(dfjan2020)
dfjanresult.to_csv("cleaned_data_jan.csv")
x={'MONTH':'JANUARY 2020','TOTAL_TRAFFIC':print_result(dfjanresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfdec2019=cleanning(dfdec2019)
print(dfdec2019)
dfdecresult=print_result_of_month(dfdec2019)
dfdecresult.to_csv("cleaned_data_dec19.csv")
x={'MONTH':'DECEMBER 2019','TOTAL_TRAFFIC':print_result(dfdecresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfnov2019=cleanning(dfnov2019)
print(dfnov2019)
dfnovresult=print_result_of_month(dfnov2019)
dfnovresult.to_csv("cleaned_data_nov19.csv")
x={'MONTH':'NOVEMBER 2019','TOTAL_TRAFFIC':print_result(dfnovresult)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfoct2019=cleanning(dfoct2019)
print(dfoct2019)
dfoct19result=print_result_of_month(dfoct2019)
dfoct19result.to_csv("cleaned_data_oct19.csv")
x={'MONTH':'OCTOBER 2019','TOTAL_TRAFFIC':print_result(dfoct19result)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
# #-----------------------------------------------------------------
dfsep2019=cleanning(dfsep2019)
print(dfsep2019)
dfsep19result=print_result_of_month(dfsep2019)
dfsep19result.to_csv("cleaned_data_sep19.csv")
x={'MONTH':'SEPTEMBER 2019','TOTAL_TRAFFIC':print_result(dfsep19result)}
dfresult=dfresult.append(x,ignore_index= True)
print(dfresult)
###################################################
dfresult.to_csv("cleaned_data.csv")
dfresult
```
